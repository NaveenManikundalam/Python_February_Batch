### Basic Plots

```python
# importing required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 50)
print (x)
```

```python
sinus = np.sin(x)
print (sinus)
```

```python
plt.plot(x, sinus)
```

```python
plt.plot(x, sinus)
plt.show()
```

```python
plt.plot(x, sinus, "o")
plt.show()
```

```python
plt.plot(x, sinus, "*")
plt.show()
```

```python
plt.plot(x, sinus, "-")
plt.show()
```

```python
plt.plot(x, sinus, "--")
plt.show()
```

```python
# Rapid multiplot
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
print (x[:5])
print (sinus[:5])
print (cosinus[:5])
```

```python
plt.plot(x, sinus, "-b", x, sinus, "ob", x, cosinus, "-r", x, cosinus, "or")
plt.xlabel("Plotting X values...")
plt.ylabel("Plotting Y values...")
plt.title("Plotting Sinus and Cosinus on X values...")
plt.show()
```

```python
plt.plot(x, sinus, "-b", x, sinus, "or", x, cosinus, "-r", x, cosinus, "ob")
plt.xlabel("Plotting X values...")
plt.ylabel("Plotting Y values...")
plt.title("Plotting Sinus and Cosinus on X values...")
plt.show()
```

```python
# Step by step plotting
plt.plot(x, sinus, label = 'sinus', color = 'blue', linestyle = "--", linewidth = 2)
plt.plot(x, cosinus, label = 'cosinus', color = 'red', linestyle = "-", linewidth = 2)
plt.legend(loc = 0)
plt.show()
```
