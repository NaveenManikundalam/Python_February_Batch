```python
# Rapid multiplot
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
print (x[:5])
print (sinus[:5])
print (cosinus[:5])
```

```python
plt.plot(x, sinus, "-b", x, sinus, "ob", x, cosinus, "-r", x, cosinus, "or")
plt.xlabel("Plotting X values...")
plt.ylabel("Plotting Y values...")
plt.title("Plotting Sinus and Cosinus on X values...")
plt.show()
```

```python
plt.plot(x, sinus, "-b", x, sinus, "or", x, cosinus, "-r", x, cosinus, "ob")
plt.xlabel("Plotting X values...")
plt.ylabel("Plotting Y values...")
plt.title("Plotting Sinus and Cosinus on X values...")
plt.show()
```

```python
# Step by step plotting
plt.plot(x, sinus, label = 'sinus', color = 'blue', linestyle = "--", linewidth = 2)
plt.plot(x, cosinus, label = 'cosinus', color = 'red', linestyle = "-", linewidth = 2)
plt.legend(loc = 0)
plt.show()
```

```python
# x ** 2 + y ** 2 = r ** 2
# y = sqrt(r ** 2 - x ** 2)
r = 10
x = np.linspace(-10, 10, 100)
y = np.sqrt(r ** 2 - x ** 2)
plt.figure(figsize = (5, 5))
plt.plot(x, y, "-r")
plt.plot(x, -y, "-r")
```

```python
# Figure size
plt.figure(figsize = (6, 5))

# Define colors and symbols
colors_edu = {"Bachelor":"red", "Master":"green", "Ph.D":"blue"}
symbols_manag = {"Y":"*", "N":"."}

# group by education x management => (3 classes x 2 classes) => 6 groups
for group, data_frame in salary_df.groupby(['education', 'management']):
    selected_education, selected_management = group
    plt.scatter(data_frame.experience, data_frame.salary, marker=symbols_manag.get(selected_management), 
                color = colors_edu.get(selected_education), s = 150, label = selected_management + " / " + selected_education)
    
plt.xlabel("Experience (In Years)...")
plt.ylabel("Salary (In INR)...")
plt.title("Experience (In Years) vs. Salary (In INR)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')

# Fixing random state for reproducibility
np.random.seed(19680801)

fig, axs = plt.subplots(ncols=2, nrows=2)
ax1, ax2, ax3, ax4 = axs.flat

# scatter plot (Note: `plt.scatter` doesn't use default colors)
x, y = np.random.normal(size=(2, 200))
ax1.plot(x, y, 'o')

# sinusoidal lines with colors from default color cycle
L = 2*np.pi
x = np.linspace(0, L)
ncolors = len(plt.rcParams['axes.prop_cycle'])
shift = np.linspace(0, L, ncolors, endpoint=False)
for s in shift:
    ax2.plot(x, np.sin(x + s), '-')
ax2.margins(0)

# bar graphs
x = np.arange(5)
y1, y2 = np.random.randint(1, 25, size=(2, 5))
width = 0.25
ax3.bar(x, y1, width)
ax3.bar(x + width, y2, width,
        color=list(plt.rcParams['axes.prop_cycle'])[2]['color'])
ax3.set_xticks(x + width, labels=['a', 'b', 'c', 'd', 'e'])

# circles with colors from default color cycle
for i, color in enumerate(plt.rcParams['axes.prop_cycle']):
    xy = np.random.normal(size=2)
    ax4.add_patch(plt.Circle(xy, radius=0.3, color=color['color']))
ax4.axis('equal')
ax4.margins(0)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```
